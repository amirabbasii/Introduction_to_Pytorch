```
import torch
from torch import optim
from torch.nn import Sequential,Linear,Sigmoid,MSELoss,Tanh
import  numpy as np
```

# Binary

preparing the dataset

```

one=torch.tensor(1)
zero=torch.tensor(0)
x=[[one,one],[one,zero],[zero,one],[zero,zero]]
y=[one,zero,zero,zero]
x=torch.tensor(x).float()
y=torch.tensor(y).float()


```

defining the model

```
model=Sequential(Linear(2,1),Sigmoid())
optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9)
```

train

```
mse=MSELoss()
for epoch in range(10000):
  pred=model(x)

  #####update
  model.train()
  optimizer.zero_grad()
  loss=mse(pred,y.reshape((4,1)))# loss
  loss.backward() #gradient 
  optimizer.step() #



  print("epoch",epoch,"     loss:",loss.item())
```

results

```
model.eval()
ans=model(x)
for i in range(4):
  print("predicted:",ans[i].item(),"   label:",y[i].item())
```
